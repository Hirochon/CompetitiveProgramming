```python
from collections import defaultdict

N = int(input())

A = list(map(int, input().split()))

ans_dict = defaultdict(int)

for i in range(N):
    if A[i] < 100:
        if A[i] < 10:
            ans_dict['0' + str(A[i])[-1:]] += 1
        else:
            ans_dict[str(A[i])] += 1
    else:
        if int(str(A[i])[-3:-2]) % 2 == 0:
            # 3桁目が偶数の時
            ans_dict[str(A[i])[-2:]] += 1
        else:
            # 3桁目が奇数の時
            ans_dict['1' + str(A[i])[-2:]] += 1

count = 0
for v in list(ans_dict.values()):
    if v > 1:
        if v == 2:
            count += 1
        else:
            value = v - 1
            count += (1 + value) * value // 2

print(int(count))
```

```python
N = int(input())
A = list(map(int, input().split()))

N = min(N, 8)
mod = [[] for _ in range(200)]

# ビットの左シフト(0001 を N個左に移動)
for S in range(1, 1 << N):
    value = 0
    for i in range(N):
        # Sのビットをiだけ右シフトして1と重なるやつを鳴らす
        if (S >> i) & 1:
            value += A[i]
            value %= 200
    mod[value].append(S)

print(mod)
    
for m in mod:
    if len(m) >= 2:
        print("Yes")
        B = m[0]
        C = m[1]
        B = [i+1 for i in range(N) if (B >> 1) & 1]
        C = [i+1 for i in range(N) if (C >> 1) & 1]
        print(len(B), *B)
        print(len(C), *C)
        break;
        
# print("No")
```

```python
import itertools
N = int(input())
A = [int(i) % 200 for i in input().split()]

mod200 = [-1] * 200
for i in itertools.product((1, 0), repeat=N):
    tmp = 0
    for j in range(N):
        tmp += A[j] * i[j]
    tmp %= 200
    if mod200[tmp] == -1:
        mod200[tmp] = i
    else:
        printx = [mod200[tmp].count(1)]
        for j, k in enumerate(mod200[tmp]):
            if k == 1:
                printx.append(j+1)
        printy = [i.count(1)]
        for j, k in enumerate(i):
            if k == 1:
                printy.append(j+1)
        if printy[0] == 0:
            break
        print("Yes")
        print(" ".join(map(str, printx)))
        print(" ".join(map(str, printy)))
        break;

print("No")
```
